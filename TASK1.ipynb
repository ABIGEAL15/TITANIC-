```python
#LabelEncoder SEX column to numerical variables
from sklearn.preprocessing import LabelEncoder
#initialize LabelEncoder
le=LabelEncoder()
#fit and transform the 'Sex' column
titanic['Sex']= le.fit_transform(titanic['Sex'])
```

```python
titanic
```

```python
#converting the AGE from float  to integer
titanic['Age']=titanic['Age'].astype(int)
```

```python
titanic
```

DIVIDING MY DATASET TO DEPENDNET AND INDEPENDNET VARIABLES

```python
from sklearn.model_selection import train_test_split
X=titanic.drop(['Survived'], axis=1)
y=titanic['Survived']

```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
X_train
```

```python
X_test
```

MODEL IMPLEMENTATION $ EVALUATION, USING BOTH THE DECISION TREE AND RANDOM FOREST  FOR PREDICTING D SURVIVAL OF PASSENGER ON THE DATASET

```python
from sklearn.tree import DecisionTreeClassifier
#train decision tree model
model=DecisionTreeClassifier()
model.fit(X_train,y_train)

```

```python
predictions=model.predict(X_test)
predictions
```

```python
#EVALUATION METRICS
from sklearn.metrics import accuracy_score
score= accuracy_score(y_test, predictions)
score
```

```python
from sklearn.ensemble import RandomForestClassifier
#train random forest model
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
random_prediction= model.predict(X_test)
random_prediction
```

```python
#EVALUATION METRICS
from sklearn.metrics import accuracy_score
score_2= accuracy_score(y_test,random_prediction)
score_2
```

```python
#comparing the two accuracy score
print(f"DecisionTree accuracy:{score : 2f}")
print(f"Random Forest accuracy:{score_2: 2f} ")
```

In conclusion, so far random forest model is greater than decision tree model, which means random forest model is the best model for this data set.
